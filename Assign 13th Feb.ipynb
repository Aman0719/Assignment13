```python
#Q1.Explain why we have to use the Exception class while creating a Custom Exception.

Ans-

When creating a custom exception in Python, it is important to inherit from the base Exception class as it has following advantages:-

1. Standardized behavior: 
By inheriting from the Exception class, our custom exception will behave like other built-in exceptions in Python. 
This means that other developers who use our code will know how to handle our exception, since they already understand how exceptions work in Python.

2. Better error messages: 
The Exception class provides methods like __str__ and __repr__ that allow us to customize the error message that will be displayed when our exception is raised. 
This can help us provide more information to developers who encounter our exception and make it easier for them to diagnose the problem.

3. Easy to catch: 
Since our custom exception inherits from the base Exception class, it can be caught in a try/except block just like any other exception. 
This makes it easy to handle our exception in a specific way and provide a fallback behavior for when things go wrong.

4. Exception hierarchy: 
In Python, there is a hierarchy of built-in exceptions that extends from the base Exception class. 
By inheriting from Exception, we can place our custom exception in the hierarchy and create a more organized structure for handling exceptions in our code.

Overall, using the Exception class as the base class for our custom exceptions is a best practice that ensures our exceptions are consistent with other exceptions in Python and easy to work with.
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans-

The LookupError class in Python is a base class for all the errors that occur when a key or index is not found in a collection or mapping object. 
It is a subclass of the Exception class and is raised when an error occurs while trying to access an element by index or key.

Using LookupError as a base class for these errors allows for more efficient and generic handling of these errors. 
Since both KeyError and IndexError are derived from LookupError, they can be caught using a single except block that handles all types of lookup errors. 
This simplifies the code and makes it more readable.

Furthermore, using LookupError as a base class allows for future changes to the exception hierarchy without breaking existing code. 
If a new type of lookup error is added to the hierarchy, it can be derived from LookupError and will be caught by any existing except blocks that handle LookupError.

Two common errors that are derived from LookupError are KeyError and IndexError.
```

```python
#1.KeyError 
#It is raised when a key is not found in a dictionary. 

#For example:
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}
print(my_dict['grape']) # KeyError: 'grape' is not a key in my_dict

"""In the above code, we are trying to access the value associated with the key 'grape' in the dictionary my_dict. However, this key is not present in the dictionary, so a KeyError is raised."""
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?

Ans-

ImportError is a Python exception that is raised when a module or package is requested to be imported, but the interpreter cannot find it. 
This error occurs when Python cannot locate the specified module or package in the current working directory or in the Python path.

For example, suppose we have a file named my_module.py that we want to import in our Python code. 
If we try to import it using the import statement, but the file is not present in the current working directory or in any of the directories specified in the Python path, an ImportError will be raised.

ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. 
It is raised when the interpreter is unable to find the specified module or package in the current working directory, in the built-in modules, or in the directories specified in the Python path.

ModuleNotFoundError was introduced to provide more detailed information about the specific module or package that could not be found. 
This allows developers to easily identify and resolve import errors by providing the name of the missing module or package.
```

```python
Q6. List down some best practices for exception handling in python.

Ans-
Here are some best practices for exception handling in Python:

1.Use specific exception types: 
Use specific exception types instead of using the generic Exception class. This helps in better understanding and handling of the errors.

2.Use try-except blocks: 
Use try-except blocks to handle exceptions in the code. This helps in catching the exception and taking appropriate action.

3.Keep the try block small:
Keep the code inside the try block small and focused. This helps in identifying the cause of the exception and taking appropriate action.

4.Avoid using bare except: 
Avoid using bare except blocks as it catches all exceptions, including those that were not intended to be caught. This makes it difficult to identify and resolve the issue.

5.Use finally block: 
Use the finally block to execute the code that needs to be executed, whether an exception was raised or not. This is useful for closing resources like files or database connections.

6.Log exceptions: 
Log the exceptions using a logging library like logging or loguru. This helps in debugging the issues and identifying the root cause.

7.Handle exceptions at the right level: 
Handle the exceptions at the appropriate level in the code. For example, if an exception occurs while accessing a database, handle the exception at the database layer instead of handling it in the application layer.

8.Reraise exceptions: 
Reraise the exceptions that cannot be handled at a particular level. This helps in propagating the exceptions to the higher level in the code where it can be handled appropriately.

9.Document the exceptions: 
Document the exceptions that can be raised by the code. This helps in better understanding and handling of the exceptions by other developers who may use the code.
```
